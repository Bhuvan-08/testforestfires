```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires.csv',header=1)
df.head()
```

```python
df.info()
```

## Data Cleaning

```python
df[df.isnull().any(axis=1)]
```

```python
# Add new column with region
df.loc[:122,'Region']=0
df.loc[122:,'Region']=1
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# Fix spaces in column names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

## Change columns as needed

```python
df.columns
```

```python
df[['month','day','year', 'Temperature', 'RH', 'Ws']] = df[['month','day','year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

## Changing to float

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.to_csv('Algerian_forest_fires_cleaned.csv',index=False)
```

# EDA

```python
df_copy = df.drop(['day','month','year'],axis=1)
```

```python
# Encoding

df_copy['Classes'] = np.where(df_copy['Classes'].astype(str).str.contains('not fire'),0,1)
```

```python
df_copy['Classes'].value_counts()
```

```python
# Density plot

plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
# Percentange for pie chart

percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
classLabels = ['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classLabels,autopct='%1.1f%%')
plt.title('Pie chart of Classes')
plt.show()
```
